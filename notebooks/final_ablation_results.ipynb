# Final Submission Results (Ablation)

This notebook loads the final ablation outputs and formats paper-ready tables.

```python
from pathlib import Path
import pandas as pd

OUT = Path('../outputs/final_ablation')
ALL = OUT / 'ablation_metrics_all.csv'
SUM = OUT / 'ablation_summary.csv'
PAIR = OUT / 'ablation_pairwise_composite.csv'

all_df = pd.read_csv(ALL)
summary_df = pd.read_csv(SUM).sort_values('composite_score', ascending=False)
pair_df = pd.read_csv(PAIR)

print('rows:', len(all_df), ' unique items:', all_df['id'].nunique())
```

## Main Results Table

```python
display(summary_df.round(4))
```

## Genre-wise Composite

```python
genre = all_df.groupby(['method','genre'])['composite_score'].mean().unstack().round(4)
display(genre)
```

## Pairwise Stability (contextual_adapt)

```python
ctx = pair_df[pair_df['method']=='contextual_adapt']['composite_score']
print('mean=', round(ctx.mean(),4), 'std=', round(ctx.std(),4), 'min=', round(ctx.min(),4), 'max=', round(ctx.max(),4))
```

## Report Paragraph

```python
best = summary_df.iloc[0]
second = summary_df.iloc[1]
n = all_df['id'].nunique()
paragraph = (
    f'On a benchmark of {n} source-target adaptation instances across six Indian subregional cultures, '
    f'the contextual adaptation method achieved the best composite score ({best.composite_score:.4f}), '
    f'outperforming lexical swap by {best.composite_score-second.composite_score:.4f}. '
    f'It also produced higher adaptation depth ({best.adaptation_depth:.4f}) with zero stereotype-risk hits in this evaluation setup.'
)
print(paragraph)
```
